# Example of cleaning

```python
import pandas as pd
```

```python
d1 = {'id':[1,2,3,4], 'name':['Alice','Bob','Chandru','John'], 'salary': ['20000','25000','50000','30000'],'city':['hyderabad','bangalore',None,'Pune']}
```

```python
df = pd.DataFrame(d1)
df.head()
```

```python
df.info()
```

1. Convert salary to integer or float
2. remove the row or set with a default value

```python
df['salary'] = df['salary'].astype('float64')
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df['city'].isna().sum()
```

```python
df_clean = df.dropna(subset=['city'])
```

```python
df_clean.info()
```
